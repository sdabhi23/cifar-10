```
import numpy as np
import tensorflow.keras as k
from datetime import datetime
```

```
(x_train, y_train), (x_test, y_test) = k.datasets.cifar10.load_data()

y_train = k.utils.to_categorical(y_train, 10)
y_test = k.utils.to_categorical(y_test, 10)
x_train = x_train / 255.
x_test = x_test / 255.
```

```
model = k.applications.MobileNetV2(input_shape=(32,32,3), include_top=True, classes=10, weights=None)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

learning_rates=[]
for i in range(5):
    learning_rates.append(2e-2)
for i in range(50-5):
    learning_rates.append(1e-2)
for i in range(100-50):
    learning_rates.append(8e-3)
for i in range(150-100):
    learning_rates.append(4e-3)
for i in range(200-150):
    learning_rates.append(2e-3)
for i in range(300-200):
    learning_rates.append(1e-3)

model_cp_path = "MobileNetV2__" + datetime.now().strftime("%Y-%m-%d-%H-%M-%S") + ".h5"

chkptr = k.callbacks.ModelCheckpoint(model_cp_path, monitor='val_accuracy', save_best_only=True)
lr_schedule = k.callbacks.LearningRateScheduler(lambda epoch: float(learning_rates[epoch]))

hist = model.fit(x_test, y_test,
                 batch_size=64,
                 epochs=300,
                 validation_data=(x_test, y_test),
                 callbacks=[lr_schedule, chkptr],
                 verbose=2)
```

```
preds = np.argmax(model.predict(x_test), axis=-1)
```

```
from sklearn.metrics import classification_report

print(classification_report(np.argmax(y_test, axis=-1), preds))
```
